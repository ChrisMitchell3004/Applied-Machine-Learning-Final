```python
import pandas as pd
df = pd.read_csv("AML Dataset.csv")
df
```

```python
inputs = df.drop('Risk_Event',axis='columns')
target = df['Risk_Event']
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Incident_ID = LabelEncoder ()
Incident_Title = LabelEncoder ()
Reoccuring_Event = LabelEncoder ()
Third_Party_Fault = LabelEncoder ()
Causal_Area = LabelEncoder ()
Impacted_Area = LabelEncoder ()
Rating = LabelEncoder ()
Initial_Loss_ = LabelEncoder ()
Recoveries = LabelEncoder ()
Net_Loss = LabelEncoder ()
Status = LabelEncoder ()
Escalated_Appropriately = LabelEncoder ()
Point_Of_Contact = LabelEncoder ()
```

```python
inputs['Incident_ID_'] = Incident_ID.fit_transform(inputs['Incident_ID'])
inputs['Incident_Title_'] = Incident_Title.fit_transform(inputs['Incident_Title'])
inputs['Reoccuring_Event_'] = Reoccuring_Event.fit_transform(inputs['Reoccuring_Event'])
inputs['Third_Party_Fault_'] = Third_Party_Fault.fit_transform(inputs['Third_Party_Fault'])
inputs['Causal_Area_'] = Causal_Area.fit_transform(inputs['Causal_Area'])
inputs['Impacted_Area_'] = Impacted_Area.fit_transform(inputs['Impacted_Area'])
inputs['Rating_'] = Rating.fit_transform(inputs['Rating'])
inputs['Initial_Loss__'] = Initial_Loss_.fit_transform(inputs['Initial_Loss_'])
inputs['Recoveries_'] = Recoveries.fit_transform(inputs['Recoveries'])
inputs['Net_Loss_'] = Net_Loss.fit_transform(inputs['Net_Loss'])
inputs['Status_'] = Status.fit_transform(inputs['Status'])
inputs['Escalated_Appropriately_'] = Escalated_Appropriately.fit_transform(inputs['Escalated_Appropriately'])
inputs['Point_Of_Contact_'] = Point_Of_Contact.fit_transform(inputs['Point_Of_Contact'])
inputs
```

```python
inputs_ = inputs.drop(['Incident_ID','Incident_Title','Reoccuring_Event','Third_Party_Fault','Causal_Area','Impacted_Area','Rating','Initial_Loss_','Recoveries','Net_Loss','Status','Escalated_Appropriately','Point_Of_Contact'],axis='columns')
inputs_
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_,target)
```

```python
model.score(inputs_,target)
```
